```python
import pandas as pd
```

```python
#united_data

united_data= pd.read_csv('../output/02_90p_autotrophic_rubisco_tree/old_auto_uclust_all_0.9.csv')
united_data = united_data[united_data.Type != 'S']
syn = pd.read_excel('milo_lab_synth_II_II.III.xlsx')
syn = syn.merge(united_data[['Query','Target']],left_on='ID',right_on='Query',how='left')
syn.loc[syn.Target.isna(),['Query','Target']] = united_data.loc[united_data.Query.str.contains('|'.join(syn[syn.Target.isna()].ID.values)),['Query','Target']].values
```

```python
run = '42_86_20190612'
results = pd.read_csv('../../../../../RO/3PG Coupled Assay/EPPS_Assays/agg_variant_data_'+run+'.csv')
res_syn = syn.merge(results,left_on='Int ID',right_on='Variant')
#res_syn[res_syn.Variant.isna()]
#res_syn[res_syn['Int ID'].isna()]
```

```python
active = res_syn[res_syn['mean norm rate [s-1]']>0.5]
active_leaves = active.groupby('Target')['mean norm rate [s-1]'].max()
lines = active_leaves.index.values+','+active_leaves.values.astype(str)+'\n'
lines = set(lines)
with open('dataset_simplebar_template.txt','r') as file:
    with open('fig3b_bar_legend2.txt', "w") as f1:
        for row in file:
            f1.write(row)
        for line in lines:
            f1.write(line)
        file.close()
        f1.close()
```

```python
fast = pd.read_csv('fast_II_rubisco.csv')
united_data['TID'] = united_data.Target.apply(lambda x: x.split(' ')[0])
united_data[united_data.TID.isin(fast.ID)].to_csv('fast_II_rubisco_nr.csv')
```

```python
f =pd.concat([active.groupby('Target')['mean norm rate [s-1]'].min(),active.groupby('Target')['mean norm rate [s-1]'].max()],axis=1)
f.columns = ['min','max']
f['diff'] = f['max'] - f['min']

f['N'] = active.groupby('Target')['mean norm rate [s-1]'].count()
f['std'] = active.groupby('Target')['mean norm rate [s-1]'].std()

f.sort_values('std',ascending=False)
```
